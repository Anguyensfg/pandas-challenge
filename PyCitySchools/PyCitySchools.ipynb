```python
# #PYCITY SCHOOLS ANALYSIS

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "raw_data/schools_complete.csv"
student_data_to_load = "raw_data/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Local Government Area Summary

# Calculate Total of Schools and Students 
school_count = len(school_data_complete_df ["school_name"].unique())
student_count = school_data_complete_df ["Student ID"].count()

# Calculate Total Budget
total_budget = school_data["budget"].sum()

# Calculate Average Scores
average_maths_score = school_data_complete_df ["maths_score"].mean()
average_reading_score = school_data_complete_df ["reading_score"].mean()


# Calculate Percentage Pass Rates at 50 or greater
passing_maths_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)].count()["student_name"]
passing_maths_percentage = passing_maths_count / float(student_count) * 100
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100

#Results not working as per example 
# overall_passing_percentage = (school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"] & \ 
                               # school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"])
    
overall_passing_percentage = (passing_maths_percentage + passing_reading_percentage) / 2


# Formating Data Set 
Local_Government_Area_Summary = pd.DataFrame({"Total Schools": [school_count], 
                                 "Total Students": [student_count], 
                                 "Total Budget": [total_budget],
                                 "Average Maths Score": [average_maths_score], 
                                 "Average Reading Score": [average_reading_score],
                                 "% Passing Maths": [passing_maths_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                 "% Overall Passing": [overall_passing_percentage]})

Local_Government_Area_Summary = Local_Government_Area_Summary[["Total Schools", "Total Students", "Total Budget",
                                     "Average Maths Score", 
                                     "Average Reading Score",
                                     "% Passing Maths",
                                     "% Passing Reading",
                                     "% Overall Passing"]]

Local_Government_Area_Summary["Total Students"] = Local_Government_Area_Summary["Total Students"].map("{:,}".format)
Local_Government_Area_Summary["Total Budget"] = Local_Government_Area_Summary["Total Budget"].map("${:,.2f}".format)

# Display Result
Local_Government_Area_Summary
```

```python
#PYCITY SCHOOLS ANALYSIS

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "raw_data/schools_complete.csv"
student_data_to_load = "raw_data/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#SCHOOL SUMMARY 

# Calculate Total of Schools and Students 
school_count = len(school_data_complete_df ["school_name"].unique())
student_count = school_data_complete_df ["Student ID"].count()

# Determine the School Type
school_types = school_data.set_index(["school_name"])["type"]

# Calculate Total Student Count
per_school_counts = school_data_complete_df["school_name"].value_counts()

# Calculate Total Budget per School 
per_school_budget = school_data_complete_df.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts

# Calculate the average test scores
per_school_maths = school_data_complete_df.groupby(["school_name"]).mean()["maths_score"]
per_school_reading = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]

# Calculate the passing scores 
school_passing_maths = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)]
school_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)]

# Calculate the passing scores by percentage
per_school_passing_maths = school_passing_maths.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

#Results not working as per example 
#per_school_overall_passing = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)].count()["school_name"] \
                             #& school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["school_name"]

overall__school_passing_percentage = (per_school_passing_maths + per_school_passing_reading) / 2


# Formating Data Set 
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": per_school_counts,
                                   "Total School Budget": per_school_budget,
                                   "Per Student Budget": per_school_capita,
                                   "Average Maths Score": per_school_maths,
                                   "Average Reading Score": per_school_reading,
                                   "% Passing Maths": per_school_passing_maths,
                                   "% Passing Reading": per_school_passing_reading,
                                   "% Overall Passing": overall__school_passing_percentage})

per_school_summary = per_school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Maths Score", "Average Reading Score", 
                                         "% Passing Maths", "% Passing Reading", 
                                         "% Overall Passing"]]
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display Result
per_school_summary
```

```python
#TOP PERFORMING SCHOOLS (BY % Overall Passing)

# Sort and display the top five performing schools by % overall passing
top_schools = per_school_summary.sort_values(["% Overall Passing"], ascending=False)
top_schools.head(5)
```

```python
#BOTTOM PERFORMING SCHOOLS (BY % Overall Passing)

# Sort and display the five worst-performing schools by % overall passing
bottom_schools = per_school_summary.sort_values(["% Overall Passing"], ascending=True)
bottom_schools.head(5)
```

```python
# Maths Scores by Year

import pandas as pd

# loading data to workspace 
school_data_to_load = "raw_data/schools_complete.csv"
student_data_to_load = "raw_data/students_complete.csv"

# Read CSV file School and student and store in workspace 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Merge data into single set 
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Create data series of scores by year level 
Year_Nine = school_data_complete_df[(school_data_complete_df["year"] == "9")]
Year_Ten = school_data_complete_df[(school_data_complete_df["year"] == "10")]
Year_Eleven = school_data_complete_df[(school_data_complete_df["year"] == "11")]
Year_Twelve = school_data_complete_df[(school_data_complete_df["year"] == "12")]

# Group year level by school name
Year_Nine_Scores = Year_Nine.groupby(["school_name"]).mean()["maths_score"]
Year_Ten_Scores = Year_Ten.groupby(["school_name"]).mean()["maths_score"]
Year_Eleven_Scores = Year_Eleven.groupby(["school_name"]).mean()["maths_score"]
Year_Twelve_Scores = Year_Twelve.groupby(["school_name"]).mean()["maths_score"]

# Combined group into data frame 
Scores_by_Year = pd.DataFrame({"Year_9": Year_Nine_Scores, "Year_10": Year_Ten_Scores,
                                "Year_11": Year_Eleven_Scores, "Year_12": Year_Twelve_Scores})

# Formating Data Set  
Scores_by_Year = Scores_by_Year[["Year_9", "Year_10", "Year_11", "Year_12"]]
Scores_by_Year.index.name = None

# Display Result
Scores_by_Year
```

```python
# Reading Score by Year

import pandas as pd

# loading data to workspace 
school_data_to_load = "raw_data/schools_complete.csv"
student_data_to_load = "raw_data/students_complete.csv"

# Read CSV file School and student and store in workspace 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Merge data into single set 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# SCORES BY SCHOOL SPENDING

# Establish the bins 
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Categorize the spending based on the bins
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)

spending_maths_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]


# ----> Check item <-----
overall_passing= (spending_passing_maths + spending_passing_reading) / 2

# Combined group into data frame 
spending_summary = pd.DataFrame({"Average Maths Score" : spending_maths_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Maths": spending_passing_maths,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing})

# Formating Data Set
spending_summary = spending_summary[["Average Maths Score", 
                                     "Average Reading Score", 
                                     "% Passing Maths", "% Passing Reading",
                                     "% Overall Passing"]]

# Display Result
spending_summary
```
